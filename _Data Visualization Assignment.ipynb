```python
#4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different
#  categories.
#Ans.
import pandas as pd
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Create synthetic dataset
categories = ['A', 'B', 'C', 'D']
data = {
    'Category': np.random.choice(categories, size=200),
    'Value': np.concatenate([
        np.random.normal(loc=20, scale=5, size=50),  # Category A
        np.random.normal(loc=35, scale=7, size=50),  # Category B
        np.random.normal(loc=50, scale=6, size=50),  # Category C
        np.random.normal(loc=65, scale=4, size=50)   # Category D
    ])
}

df = pd.DataFrame(data)

# Visualize the distribution using a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='Set2')
plt.title('Distribution of Numerical Values Across Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(True, axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
#Ans.
# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
n_samples = 100
feature_1 = np.random.normal(0, 1, n_samples)
feature_2 = feature_1 + np.random.normal(0, 0.1, n_samples)  # Highly correlated with feature_1
feature_3 = 0.5 * feature_1 - 0.2 * feature_2 + np.random.normal(0, 0.2, n_samples)  # Moderately correlated
feature_4 = np.random.normal(0, 1, n_samples)  # Not correlated

# Create DataFrame
df = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Feature 3': feature_3,
    'Feature 4': feature_4
})

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.tight_layout()
plt.show()
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function 
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
 'X': x.flatten(),
 'Y': y.flatten(),
 'Z': z.flatten()
}
df = pd.DataFrame(data)

#Ans.
# Plotting
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Reshape for surface plot
X = df['X'].values.reshape(100, 100)
Y = df['Y'].values.reshape(100, 100)
Z = df['Z'].values.reshape(100, 100)

# Create surface plot
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')

# Add labels and colorbar
ax.set_title('3D Surface Plot of $z = \sin(\sqrt{x^2 + y^2})$')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

plt.show()
```

```python
#1. Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.
#Ans.
from bokeh.plotting import figure, show, output_file

# Generate data
x = np.linspace(0, 10, 500)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add a line renderer
p.line(x, y, line_width=2, color='blue', legend_label="sin(x)")

# Show the results in a browser
output_file("sine_wave.html")
show(p)
```

```python
#2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the
#   'sizes' and 'colors' columns.
#Ans.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Display Bokeh plots in the notebook
output_notebook()

# Generate random data
n = 100
x = np.random.rand(n) * 100
y = np.random.rand(n) * 100
sizes = np.random.rand(n) * 30 + 10  # sizes between 10 and 40
colors = [f"#{int(r):02x}{int(g):02x}{int(b):02x}" 
          for r, g, b in zip(np.random.rand(n)*255, np.random.rand(n)*255, np.random.rand(n)*255)]

# Create a data source
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create the plot
p = figure(title="Random Bokeh Scatter Plot", tools="pan,wheel_zoom,reset", width=600, height=400)
p.scatter('x', 'y', size='sizes', color='colors', alpha=0.6, source=source)

# Show the plot
show(p)
```

```python
#3.  Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
#Ans.
from bokeh.io import show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource
from bokeh.transform import dodge

# Enable output in notebook (optional for Jupyter)
output_notebook()

# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create data source
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create figure
p = figure(x_range=fruits, height=350, title="Fruit Counts",
           toolbar_location=None, tools="")

# Add vertical bars
p.vbar(x='fruits', top='counts', width=0.6, source=source, legend_field="fruits",
       line_color='white', fill_color='navy')

# Customize plot
p.y_range.start = 0
p.xgrid.grid_line_color = None
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show plot
show(p)
```

```python
#4.  Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
#Ans.
from bokeh.plotting import figure, show, output_notebook

# Output to notebook (or use output_file("histogram.html") to save it as HTML)
output_notebook()

# Create the figure
p = figure(title="Histogram of Normally Distributed Data",
           background_fill_color="#fafafa",
           x_axis_label='Value', y_axis_label='Frequency')

# Add quad glyphs to represent histogram bars
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],
       fill_color="navy", line_color="white", alpha=0.7)

# Show the plot
show(p)
```

```python
#5. Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
#Ans.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.layouts import layout
from bokeh.transform import linear_cmap

# Output to notebook
output_notebook()

# Set up the color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Create figure
p = figure(title="Heatmap Example", x_range=(0, 1), y_range=(0, 1),
           x_axis_label='X', y_axis_label='Y',
           tools='pan,box_zoom,reset,save')

# Display image (transposed because image() expects shape as [rows, cols])
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
